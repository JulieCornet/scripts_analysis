```python
from re import split
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import glob
import re
from matplotlib import rc
from scipy.optimize import curve_fit
from scipy.special import legendre
from ipywidgets import *
pi = np.pi
```

```python
fname = '/users/cornet/Documents/PhD/Matthieu/POPC_tension/calculate-bilayer-power-spectrum-master/output_test/output-0008-undulation.dat'
l1 = []
l2 = []
l3 = []
with open(fname,'r') as fi:
    data=fi.readlines()
        
for i,line in enumerate(data):
    spl=split(' |\n',line)
    l1.append(float(spl[0]))
    l2.append(float(spl[1]))
    l3.append(float(spl[2]))
```

```python
l1 = np.array(l1)
l2 = np.array(l2)
l3 = np.array(l3)
```

```python
fig, ax=plt.subplots(figsize=(12,8))

ax.plot(np.log10(l1),l2*l1**4)

plt.xlabel('q ($nm^{⁻1}$) ?',fontsize=22)
plt.ylabel('$q⁴$ intensity ($nm^2$) ?',fontsize=22)

plt.xlim(np.log10(0.1),np.log10(3.0))
#plt.ylim(-0.15,0.22)
#plt.grid(True)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.savefig('/users/cornet/Documents/PhD/Matthieu/POPC_tension/calculate-bilayer-power-spectrum-master/power_spectra_loglin.pdf', format='pdf',bbox_inches='tight')
```

```python
fig, ax=plt.subplots(figsize=(12,8))

ax.loglog(l1,l2*l1**4)

plt.xlabel('q ($nm^{⁻1}$) ?',fontsize=22)
plt.ylabel('$q⁴$ intensity ($nm^2$) ?',fontsize=22)

plt.xlim(0.1,3.0)
#plt.ylim(-0.15,0.22)
#plt.grid(True)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.savefig('/users/cornet/Documents/PhD/Matthieu/POPC_tension/calculate-bilayer-power-spectrum-master/power_spectra_loglog.pdf', format='pdf',bbox_inches='tight')
```
